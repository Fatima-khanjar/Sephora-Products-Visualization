**Comprehensive Analysis of Sephora Product Trends and Consumer Engagement**

**Introduction**

This dataset contains detailed information about a variety of products available from Sephora.Below is a breakdown of the columns included in this dataset:

1. product_id: A unique identifier for each product, which aids in tracking and analysis.
2. product_name: The name of the product, which often reflects its key benefits or ingredients.
3. brand_id: A unique identifier for each brand, useful for brand-specific analyses.
4. brand_name: The name of the brand that manufactures the product, important for understanding brand influence and loyalty.
5. loves_count: A count of how many times a product has been 'loved' by customers, an indicator of popularity.
6. rating: The average consumer rating of the product, which reflects customer satisfaction.
7. reviews: The number of reviews a product has received, providing insight into consumer engagement.
8. size: The size of the product, relevant for price comparison and value assessment.
9. variation_type: Describes the type of variations available for the product, such as size or shade.
10. variation_value: The specific value of the variation, providing further detail on product options.:
11. Ingredients:List of ingredients used in the product, crucial for consumers who are concerned about allergies, product safety, and the environmental impact of their cosmetics.
12. price: The retail price of the product, important for economic analysis and market positioning.
13. Limited_edition:Indicates whether the product is a limited edition item. Limited edition products are often produced in smaller quantities for a short period of time, making them desirable for collectors and fans of exclusive items.
14. new:Flags whether the product is a new addition to the Sephora lineup. This is important for tracking recent market trends and consumer interest in the latest beauty products.
15. online_only: Indicates whether the product is exclusively sold online, highlighting distribution strategies.
16. out_of_stock: Whether the product is currently out of stock, which can signal demand issues or supply chain disruptions.
17. sephora_exclusive: Whether the product is exclusively sold at Sephora, pointing to exclusivity agreements.
18. highlights: Key features or selling points of the product, essential for marketing analysis.
19. primary_category, secondary_category, tertiary_category: These categorize the products into various segments, useful for market segmentation analysis.
20. child_count, child_max_price, child_min_price: Information about product variants, offering insights into the range of options and pricing flexibility.

**Goals of Analysis**

The objective of this analysis is to delve deep into the dataset to discover trends and insights that could guide strategic business decisions. Specifically, the analysis will focus on:

*  Identifying the most popular and highly rated products and brands.
*  Examining product availability and distribution strategies based on online exclusivity and stock status.
* Analyzing market segmentation based on the product categories and price points.
*  Understanding customer engagement through reviews and love counts.

## **3. Preparing Data for Visualization**

## Step 1: Remove Duplicate Rows

```python
df.drop_duplicates(inplace = True)
```

## Step 2: Hnadling Missing Values

```python
# Fill missing values
df['reviews'].fillna(df['reviews'].median(), inplace=True)
df['rating'].fillna(df['rating'].median(), inplace=True)
df['size'].fillna('Not specified', inplace=True)
df['variation_type'].fillna('Unknown', inplace=True)
df['variation_value'].fillna('Unknown', inplace=True)
df['variation_desc'].fillna('Unknown', inplace=True)
df['ingredients'].fillna('Unknown', inplace=True)
df['value_price_usd'].fillna(0, inplace=True)
df['sale_price_usd'].fillna(0, inplace=True)
df['highlights'].fillna('Unknown', inplace=True)
df['primary_category'].fillna('Unknown', inplace=True)
df['secondary_category'].fillna('Unknown', inplace=True)
df['tertiary_category'].fillna('Unknown', inplace=True)
df['child_max_price'].fillna(df['child_max_price'].median(), inplace=True)
df['child_min_price'].fillna(df['child_min_price'].median(), inplace=True)


print(df.isnull().sum())
```

## Step 3: Strip Whitespace from Strings

```python
# Strip whitespace from string columns
df['product_name'] = df['product_name'].str.strip()
df['brand_name'] = df['brand_name'].str.strip()
df['ingredients'] = df['ingredients'].str.strip()
df['variation_type'] = df['variation_type'].str.strip()
df['variation_desc'] = df['variation_desc'].str.strip()
df['variation_value'] = df['variation_value'].str.strip()
df['primary_category'] = df['primary_category'].str.strip()
df['secondary_category'] = df['secondary_category'].str.strip()
df['tertiary_category'] = df['tertiary_category'].str.strip()
df['highlights'] = df['highlights'].str.strip()
```

## Step 4: Convert Columns to Appropriate Data Types

```python
# Convert `limited_edition`, `new`, `online_only`, `out_of_stock`, and `sephora_exclusive` to boolean
df['limited_edition'] = df['limited_edition'].astype(bool)
df['new'] = df['new'].astype(bool)
df['online_only'] = df['online_only'].astype(bool)
df['out_of_stock'] = df['out_of_stock'].astype(bool)
df['sephora_exclusive'] = df['sephora_exclusive'].astype(bool)

# Convert `loves_count`, `price_usd` to numeric
df['loves_count'] = pd.to_numeric(df['loves_count'], errors='coerce')
df['price_usd'] = pd.to_numeric(df['price_usd'], errors='coerce')
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['reviews'] = pd.to_numeric(df['reviews'], errors='coerce')

```

## Step 5: Handling the Outliers

```python
# Handle outliers
upper_limit_loves_count = df['loves_count'].quantile(0.95)
df['loves_count'] = np.where(df['loves_count'] > upper_limit_loves_count, upper_limit_loves_count, df['loves_count'])

upper_limit_reviews = df['reviews'].quantile(0.95)
df['reviews'] = np.where(df['reviews'] > upper_limit_reviews, upper_limit_reviews, df['reviews'])
```

## Step 6: Deriving New Columns


```python
# New features
#The discounted column indicates whether a product is sold at a discount
df['discounted'] = df['sale_price_usd'] < df['price_usd']
# Calculate the discount percentage for each product
df['discount_percentage'] = (df['price_usd'] - df['sale_price_usd']) / df['price_usd'] * 100
df['well_received'] = df['rating'] > 4.0  # Assuming 4.0 as a threshold for good ratings
```

## Step 7: Normalization/Standarization


```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

scaler = MinMaxScaler()
df[['price_usd', 'loves_count', 'reviews']] = scaler.fit_transform(df[['price_usd', 'loves_count', 'reviews']])
```

## 4.3 Product Availability


**Purpose**: These pie charts display the proportion of products that are online only, exclusive to Sephora, and out of stock. Understanding these aspects can help in inventory and channel management.

**Insight**: High percentages of exclusive or online-only products might suggest a niche market strategy or an online sales focus. Products frequently out of stock could indicate supply chain issues or high demand that isn't being met.


```python
def plot_pie(column, title):
    counts = df[column].value_counts()
    plt.figure(figsize=(7, 7))
    plt.pie (counts, labels=counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('light:b'))
    plt.title(title)
    plt.show()

plot_pie('online_only', 'Proportion of Online Only Products')
plot_pie('sephora_exclusive', 'Proportion of Sephora Exclusive Products')
plot_pie('out_of_stock', 'Proportion of Out of Stock Products')
```

**Proportion of Online Only Products**:

21.9% of products are available online only, emphasizing the importance of e-commerce for these products.

78.1% of products are available both online and in physical stores, showing a balanced distribution between online and in-store availability.

" This data emphasizes the need for a strong online presence in addition to physical stores."


**Proportion of Sephora Exclusive Products**:

27.9% of products are exclusive to Sephora, highlighting the retailer's strategy to offer unique products.

72.1% of products are not exclusive, indicating a diverse range of offerings available elsewhere.

" This insight can help in strategizing product placement and partnerships."



**Proportion of Out of Stock Products**:

7.4% of products are out of stock, suggesting a moderate level of demand or potential supply chain issues.

92.6% of products are in stock, indicating good availability for most products

## 4.9 Brand Diversity in Top Categories



**Purpose**: To see which brands dominate the top product categories.

**Insight**: This can help understand brand concentration and competitiveness within key product areas.

```python
# Find the top 3 most populated categories
top_categories = df['primary_category'].value_counts().nlargest(3).index

# Filter data for these top categories
top_category_data = df[df['primary_category'].isin(top_categories)]

# Identify the top brands in each category
top_brands_per_category = top_category_data.groupby('primary_category')['brand_name'].apply(lambda x: x.value_counts().nlargest(5).index)

# Flatten the list of top brands
top_brands = top_brands_per_category.explode().unique()

# Filter data for the top brands in the top categories
top_brands_data = top_category_data[top_category_data['brand_name'].isin(top_brands)]

# Plot
plt.figure(figsize=(14, 8))
sns.countplot(x='primary_category', hue='brand_name', data=top_brands_data, palette='light:b')
plt.title('Top Brands in Each of the Top Categories')
plt.xlabel('Primary Category')
plt.ylabel('Number of Products')
plt.xticks(rotation=45)
plt.legend(title='Brand', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

**Makeup**:
One brand (possibly Fenty Beauty by Rihanna or SEPHORA COLLECTION) dominates with over 175 products.Other top brands in the Makeup category have fewer products, with counts around 50 to 75.

**Skincare:**
Multiple brands have a significant number of products, ranging from around 25 to 100.The distribution is more even compared to the Makeup category, indicating a competitive market with several leading brands.

**Hair:**
The Hair category also shows several brands with a significant number of products.The counts range from around 25 to 75, indicating multiple brands with a strong presence in this category.
